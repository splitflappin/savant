```python
import pybaseball as pyb
```

```python
# pulls data from baseball savant
def savData(start_date, end_date):
    # format = YYYY-MM-DD
    sav = pyb.statcast(start_dt=start_date, end_dt=end_date)
    return sav
```

```python
sav25 = savData("2025-01-01", "2025-11-13")
```

```python
sav24 = savData("2024-01-01", "2024-12-31")
```

```python
sav25.to_csv("data/sav25.csv", index=False)
```

```python
sav24.to_csv("data/sav24.csv", index=False)
```
